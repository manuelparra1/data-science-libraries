## Line charts

Limits - plt.xlim, plt.ylim  
Color - c argument  
Alpha - transparency  
Multiple plots  

```python
x = np.array(range(1, 11))

x
```

```python
y = x**2

y
```

```python
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlim(5,10)
```

```python
plt.plot(x,y)
plt.xlim(5,10)
plt.ylim(60,100)
```

```python
plt.plot(x,y, c='rebeccapurple')
plt.xlim(5,10)
plt.ylim(60,100)
```

```python
z=x**2-4
```

```python
plt.plot(x,y, c='oldlace')
plt.plot(x,z, c='firebrick', alpha=1)
plt.xlim(9,10)
plt.ylim(90,100)
```

## Labels, Titles, and Axes

plt.title  
plt.xlabel  
plt.ylabel  

Write LaTeX symbols by wrapping text with dollar signs ($)

```python
plt.scatter(x,y, c='olive',s=1000)
plt.scatter(x,z, c='darkred',s=100)
plt.title('Olive Pits = $sin(x)$',fontsize=35)
plt.xlabel('My xvalues.')
plt.ylabel('My yvalues.')
```

## Custom Ticks

plt.xticks  
plt.yticks  

```python
plt.scatter(x,y, c='olive',s=1000)
plt.scatter(x,z, c='darkred',s=100)
plt.title('Olive Pits = $ \Omega $',fontsize=35)
plt.xlabel('My xvalues.')
plt.ylabel('My yvalues.')
plt.xticks([1,5,10])
plt.yticks([20,40,80,100])
```

## Custom Line Types

Line style - ls argument (accepts ":" and "- -")

```python
plt.plot(x,y, c='olive',ls='--')
plt.scatter(x,z, c='green',s=400)
plt.scatter(x,z-2, c='darkred',s=100)
plt.scatter(x,z-3, c='white',s=10)

plt.title('Christmas Lights = $ \Omega $',fontsize=35)
plt.xlabel('My xvalues.')
plt.ylabel('My yvalues.')
plt.xticks([1,5,10])
plt.yticks([20,40,80,100])
```

## Adding text

plt.text - x and y coordinates, text, fontsize, color

plt.annotate - text, pointer coordinates, text coordinates, arrow properties

```python
plt.plot(x,y, c='olive',ls='--')
plt.scatter(x,z, c='green',s=400)
plt.scatter(x,z-2, c='darkred',s=100)
plt.scatter(x,z-3, c='white',s=10)

plt.title('Christmas Lights = $ \Omega $',fontsize=35)
plt.xlabel('My xvalues.')
plt.ylabel('My yvalues.')
plt.xticks([1,5,10])
plt.yticks([20,40,80,100])
plt.text(6,20,'Merry Christmas',fontsize=15, c='firebrick')
```

```python
plt.plot(x,y, c='olive',ls='--')
plt.scatter(x,z, c='green',s=400)
plt.scatter(x,z-2, c='darkred',s=100)
plt.scatter(x,z-3, c='white',s=10)

plt.title('Christmas Lights = $ \Omega $',fontsize=35)
plt.xlabel('My xvalues.')
plt.ylabel('My yvalues.')
plt.xticks([1,5,10])
plt.yticks([20,40,80,100])
plt.text(6,20,'Merry Christmas',fontsize=15, c='firebrick')
plt.annotate('Arrow', xy=(6, 40), xytext=(5, 80),
             arrowprops={'facecolor': 'blue'})
```

## Figure size and legends

plt.figure - figsize argument to determine size of plot  
plt.legend - loc argument for location

```python
plt.figure(figsize=(6,4),dpi=120)
plt.plot(x,y, c='olive',ls='--')
plt.scatter(x,z, c='green',s=400)
plt.scatter(x,z-2, c='darkred',s=100)
plt.scatter(x,z-3, c='white',s=10)

plt.title('Christmas Lights = $ \Omega $',fontsize=35)
plt.xlabel('My xvalues.')
plt.ylabel('My yvalues.')
plt.xticks([1,5,10])
plt.yticks([20,40,80,100])
plt.text(6,20,'Merry Christmas',fontsize=15, c='firebrick')
plt.annotate('Arrow', xy=(6, 40), xytext=(5, 80),
             arrowprops={'facecolor': 'blue'})
plt.legend(['Wire','Light','Light','Light'])
```

## Saving figures

plt.savefig - input the filename, default .png filetype

```python
plt.figure(figsize=(6,4),dpi=120)
plt.plot(x,y, c='olive',ls='--')
plt.scatter(x,z, c='green',s=400)
plt.scatter(x,z-2, c='darkred',s=100)
plt.scatter(x,z-3, c='white',s=10)

plt.title('Christmas Lights = $ \Omega $',fontsize=35)
plt.xlabel('My xvalues.')
plt.ylabel('My yvalues.')
plt.xticks([1,5,10])
plt.yticks([20,40,80,100])
plt.text(6,20,'Merry Christmas',fontsize=15, c='firebrick')
plt.annotate('Arrow', xy=(6, 40), xytext=(5, 80),
             arrowprops={'facecolor': 'blue'})
plt.legend(['Wire','Light','Light','Light'])
plt.savefig('christmas_scatter_plot')
```

## Subplots

plt.subplot - number of rows, number of columns, selected figure  
plt.subplot - shorthand notation  
plt.suptitle - title for collection of subplots

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Plot 1')

plt.subplot(1,2,2)
plt.plot(x,z,c='firebrick')
plt.title('Plot 2')

plt.suptitle('Comparison')
```
