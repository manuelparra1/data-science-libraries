```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

my_first_dataframe = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
#    a. Create a column named passing_english that indicates whether 
#       each student has a passing grade in english.
```

```python
my_first_dataframe['passing_english'] = my_first_dataframe.english >= 70
my_first_dataframe
```

```python
#    b. Sort the english grades by the passing_english column. 
#       How are duplicates handled?

my_first_dataframe.sort_values(by='passing_english')
```

```python
#    c. Sort the english grades first by passing_english and then
#       by student name. All the students that are failing english
#       should be first, and within the students that are failing
#       english they should be ordered alphabetically. The same should
#       be true for the students passing english. (Hint: you can pass 
#       a list to the .sort_values method)

# english by english descending, [english<70+name alpha], [english>70+name alpha]
# students by name


my_first_dataframe.sort_values(by=['passing_english', 'name'])
```

```python
#    d. Sort the english grades first by passing_english, and then by the 
#       actual english grade, similar to how we did in the last step.
my_first_dataframe.sort_values(by=['passing_english', 'english'])
```

```python
#    e. Calculate each students overall grade and add it as a column on 
#       the dataframe. The overall grade is the average of the math, 
#       english, and reading grades.
my_first_dataframe.sort_values(by=['passing_english', 'english'])

my_first_dataframe['overall_grade'] = my_first_dataframe[["math","english","reading"]].mean(axis=1)

my_first_dataframe
```

```python
#    f. How many animals are hoppers that are above the median speed? What percentage is this?
hoppers = mammals_data[mammals_data['hoppers']]

median_speed = mammals_data['speed'].median()
print(median_speed)
```
